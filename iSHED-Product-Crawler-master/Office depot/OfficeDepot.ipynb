```python
def Get_product(product,url):
    time.sleep(1)
    product = product
    url = url
    base_url     = 'https://www.officedepot.com'
    complete_url = base_url + url 
#     print(complete_url)
    result       = requests.get(complete_url,timeout =None)
    soap =BeautifulSoup(result.content,'lxml')
    product_details = {}
    for span in soap.find_all('div',{'class':'unified_price_row unified_sale_price red_price'}):
        try:
            product_details['special_price']=span.find('span').text
        except:
            product_details['special_price'] = 'Nill'
#     for div in soap.find_all('div',{'class':'unified_price_row unified_reg_price'}):
#         try:
#             product_details['regular_price'] = div.find(class_ ='strike').text.strip()
#         except:
#             product_details['regular_price'] ='Nill'
    for para in soap.find_all('div',class_='sku_desc show'):
        try:
            product_details['Description'] =para.find('p').text.strip()
            print(product_details['Description'])
        except:
            product_details['Description'] = 'Nill'
    for div in soap.find_all('div',{'id':'productPurchase'}):
        try:
            product_details['orgial_price'] = div.find(class_='price_column right').text.strip()
        except:
            product_details['orgial_price'] = 'Nill'
    try:
        product_details['image'] = soap.find('img',{'itemprop':'image'})['src']
    except:
        product_details['image'] = 'Nill'
    try:
        product_details['item_number'] = soap.find('td',{'id':'basicInfoCustomerSku'}).text.strip()
    except:
        product_details['item_number'] = 'Nill'
    
    try:
        product_details['Brand'] = soap.find('td',{'id':'attributebrand_namekey','itemprop':'brand'}).text.strip()
    except:
        product_details['Brand'] = 'Nill'
    try:
        Name = soap.find('h1',{'itemprop':'name'}).text.split()
        product_details['Sub-category'] = ' '.join(Name[:-3])
    except:
        product_details['Sub-category'] = 'Nill'
    try:
        product_details['Model_number'] = soap.find('td',{'id':'attributemodelkey','itemprop':'model'}).text.strip()
    except:
        product_details['Model_number'] = 'Nill'
    product_details['Category'] = 'Nill'
    print(product_details['Brand'])
#     print(product_details)
    product_csv_creator_file(product_details,product)

    
```

```python
import csv
import os
import os.path
import io
import datetime
today_date = datetime.date.today()


def product_csv_creator_file(data, category_name):

    if not os.path.exists('product_csv/'+str(today_date)+'/'):
        os.makedirs('product_csv/'+str(today_date)+'/')
    filename = 'product_csv/'+str(today_date)+'/'+category_name+'.csv'
    file_exists = os.path.isfile(filename)
    with io.open(filename, 'a', encoding='utf-8') as csvfile:
        fieldnames = ['Category','Sub-category' ,'Orginal_price','special_price',
                      'Item_number','Brand','image','Model_number','Description']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        if not file_exists:
            writer.writeheader()
        try:    
            writer.writerows([{'Category':data['Category'],'Sub-category':data['Sub-category'],'Orginal_price':data['orgial_price']
                           ,'special_price':data['special_price'],'Item_number':data['item_number'], 'Brand':data['Brand'],
                           'image':data['image'],'Model_number':data['Model_number'],
                           'Description':data['Description'],
                          }])
        except:
            writer.writerows([{'Category':data['Category'],'Sub-category':data['Sub-category'],'Orginal_price':'Nill'
                           ,'special_price':'Nill','Item_number':data['item_number'], 'Brand':data['Brand'],
                           'image':data['image'],'Model_number':data['Model_number'],
                           'Description': 'Nill',
                          }])
        print("Writing complete")
```
